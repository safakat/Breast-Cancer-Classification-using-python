<h3><b>Model Training</b></h3>

```python
X = cancer_df.drop(['target'],axis=1)
```

```python
X.head()
```

```python
y = cancer_df['target']
```

```python
y.head()
```

```python
X.shape
```

```python
y.shape
```

```python
#import libraries
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train
```

```python
X_test
```

```python
y_test
```

```python
y_train
```

```python
#import libraries
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train,y_train)
```

<h3><b>Evaluating The Model</b></h3>

```python
y_predict = svc_model.predict(X_test)
```

```python
y_predict
```

```python
y_test
```

```python
cm = confusion_matrix(y_test,y_predict)
```

```python
sn.heatmap(cm,annot=True)
```

<h3><b>Improving The Model.</h3></b>

```python
min_train = X_train.min()
min_train
```

```python
range_train = (X_train - min_train).max()
range_train
```

```python
X_train_scaled = (X_train - min_train)/range_train
```

```python
X_train_scaled
```

```python
sn.scatterplot(x = X_train['mean area'],y = X_train['mean smoothness'],hue=y_train)
```

```python
sn.scatterplot(x = X_train_scaled['mean area'],y = X_train_scaled['mean smoothness'],hue=y_train)
```

```python
min_test = X_test.min()
range_test = (X_test - min_test).max()
X_test_scaled = (X_test - min_test)/range_test
```

```python
svc_model.fit(X_train_scaled,y_train)
```

```python
y_predict1=svc_model.predict(X_test_scaled)
```

```python
cm1 = confusion_matrix(y_test,y_predict1)
```

```python
sn.heatmap(cm1, annot = True)
```
